### Esercizio 12.1

#### Svolgimento

Data una rete profonda deputata al riconoscimento delle dieci cifre scritte a mano, l'esercizio richiede di valutare *accuracy* e *loss* dei dati di *training* e di *test* per la stessa rete utilizzando due ottimizzatori diversi.

Nella prima rete si utilizza l'ottimizzatore *Adam* e nella seconda *Nadam*. L'ottimizzatore *Adam* è un metodo di *stochastic gradient descent* basato su una stima adattiva ai parametri. Il *Nadam* è simile al precedente e utilizza il momento di Nesterov. Tali ottimizzatori accelerano il processo di ricerca, perché permettono di superare facilmente zone in cui la funzione costo, ossia la *loss*, ha un andamento piatto. 

Nella fase di *training* si allena la prima rete per $5$ epoche e la seconda per $10$ epoche mescolando i dati con l'opzione *shuffle*.

#### Risultati

Di seguito i codici di creazione, compilazione e allenamento delle due reti neurali profonde, i grafici di *accuracy* e *loss* e i risultati di predizione di alcune cifre del dataset MNIST.

Prima DNN:

```python
import tensorflow as tf
from tensorflow import keras
import os
import numpy as np
import matplotlib.pyplot as plt
seed=0
np.random.seed(seed) # fix random seed
tf.random.set_seed(seed)
```

```python
from keras.datasets import mnist

# input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 
# output
num_classes = 10 # 10 digits

# the data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print()
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
# reshape data, it could depend on Keras backend
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

# cast floats to single precision
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# rescale data in interval [0,1]
X_train /= 255
X_test /= 255

# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

def create_DNN():
    # instantiate model
    model = Sequential()
    # add a dense all-to-all relu layer
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='relu'))
    # add a dense all-to-all relu layer
    model.add(Dense(100, activation='relu'))
    # apply dropout with rate 0.5
    model.add(Dropout(0.5))
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    
    return model

print('Model architecture created successfully!')
```

```python
from keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam
#Uso Adam
def compile_model():
    # create the model
    model=create_DNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=Adam(),
                  metrics=['acc'])
    return model

print('Model compiled successfully and ready to be trained.')
```

```python
# training parameters
batch_size = 32
epochs = 5

# create the deep neural net
model_DNN = compile_model()

# train DNN and store training info in history
history = model_DNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
# evaluate model
score = model_DNN.evaluate(X_test, Y_test, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history

# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.ylabel('model accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()
```

Si osserva che con *Adam* come ottimizzatore si ha per i dati di *test* un'*accuracy*, ossia una percentuale di successo nella classificazione dei dati, dell'$98\%$ e una *loss*, funzione da minimizzare settando opportunamente pesi e *bias*, di *0.08*. Da questi valori si può osservare che la rete sembra essere più performante di quella ottimizzata con *stochastic gradient descent* presente nelle consegne di quest'esercitazione.

```python
#X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)
predictions = model_DNN.predict(X_test)

X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)

plt.figure(figsize=(15, 15)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nPredicted:    {}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))    
    plt.axis('off') 
plt.show()
```

La rete opera una predizione corretta per le dieci cifre sopra riportate.

### Esercizio 12.2

#### Svolgimento

Si cambia ora la rete profonda precedente in una rete di tipo convoluzionale, con ottimizzatore *Nadam*, allenata per $10$ epoche. Con una rete convoluzionale si garantisce l'invarianza traslazionale dei dati che essa deve riconoscere.

La rete è composta da un primo *layer* che svolge l'operazione di convoluzione con sei filtri $5$x$5$ e con *stride*, ossia numero di neuroni di cui viene traslato il filtro nella convoluzione, $1$. Il secondo è un *layer* di *pooling*. Esso riduce le dimensione dell'*output* sostituendo ad una regione di neuroni $2$x$2$ un solo neurone di *output* pari a quello massimo di quella regione. Il terzo strato è un *layer* convoluzionale di $16$ filtri $5$x$5$ e *stride* $1.$ Gli strati successivi sono un *max pooling* di *stride* $2$, un *dropout*, uno strato di *flatten* che trasforma l'*input* in un vettore e l'ultimo *layer*, di tipo *dense* e attivato da *softmax*, per avere in uscita il risultato del neurone più acceso, al cui *output* è associata probabilità maggiore. I *layer* convoluzionali sono invece attivati dalla funzione tangente iperbolica.

#### Risultati

Dopo aver riadattato i dati di *train* e *test* per poterli sfruttare nella rete convoluzionale, si riportano i codici di creazione e allenamento della rete e si valutano le sue performance.

```python
# you will need the following for Convolutional Neural Networks
from keras.layers import Flatten, Conv2D, MaxPooling2D

# reshape data, depending on Keras backend
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)
    
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print()
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
def create_CNN():
    # instantiate model
    model = Sequential()
    # add first convolutional layer with 6 filters
    model.add(Conv2D(6, kernel_size=(5, 5), strides=1, activation='tanh', input_shape=input_shape))
    model.add(MaxPooling2D(pool_size=(2, 2), strides=2))
    model.add(Conv2D(16, kernel_size=(5, 5), strides=1, activation='tanh'))
    model.add(MaxPooling2D(strides=2))
    # apply dropout with rate 0.5
    model.add(Dropout(0.5))
    model.add(Flatten())
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    return model

print('Model architecture created successfully!')
```

```python
def compile_model():
    # create the model
    model=create_CNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=Nadam(),
                  metrics=['acc'])
    return model

print('Model compiled successfully and ready to be trained.')
```

```python
# training parameters
batch_size = 32
epochs = 10

# create and compile the deep neural net
model_CNN = compile_model()

# train CNN and store training info in history
history = model_CNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          shuffle=True,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
# evaluate model
score = model_CNN.evaluate(X_test, Y_test, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history

# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.ylabel('model accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()
```

L'*accuracy* sui dati di *test* è paragonabile a quella delle reti profonde precedenti. Rispetto alle reti neurali profonde, una rete di tipo convoluzionale si caratterizza per la sua capacità di sfruttare le simmetrie tipiche dell'oggetto che deve riconoscere, come le cifre in questo caso, per una classificazione indipendente dalla posizione dell'oggetto all'interno dell'immagine.

```python
#X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)
predictions = model_CNN.predict(X_test)

X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)

plt.figure(figsize=(15, 15)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nPredicted:    {}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))    
    plt.axis('off') 
plt.show()
```

Tutte le cifre sopra riportate vengono riconosciute dalla CNN.
