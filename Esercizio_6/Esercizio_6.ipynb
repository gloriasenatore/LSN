### Esercizio 6.1

#### Svolgimento

L'esercizio riguarda la simulazione del modello di Ising unidimensionale. Si ha un sistema di $50$ particelle che possono avere spin up o down, con condizioni al bordo periodiche. La dinamica del sistema è descritta dalla seguente hamiltoniana di interazione a primi vicini 
$$ H = -J\sum_{i=1}^N s_i s_{i+1} - \frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) $$
con $J$ coefficiente di accoppiamento e $h$ campo esterno moltiplicato per un momento magnetico, come il magnetone di Bohr $\mu_{b}$ ad esempio.

Viene campionato il peso di Boltzmann $p(\{s_i\}) = \exp(-\beta H)/Z$ con gli algoritmi di Metropolis e di Gibbs. Un passo dell'algoritmo Monte Carlo consiste nel scegliere casualmente uno dei $50$ spin per $50$ volte. 

Nel caso di Metropolis, per ogni spin viene valutata la sua stessa energia, $E_{old} = -J s_{i} (s_{i-1} + s_{i+1}) - h s_{i}$, e la sua energia dopo esser stato cambiato da up a down o viceversa, $E_{new}$, con tutti gli altri spin rimasti invariati, in modo che la probabilità di transizione sia simmetrica, $T(old|new) = T(new|old)$. La matrice di accettanza dell'algoritmo di Metropolis è:
$$A(new|old) = min(1, e^{-\beta(E_{new}-E_{old})})$$
Lo spin viene di sicuro cambiato se $E_{new} < E_{old}$ oppure con probabilità $e^{-\beta(E_{new}-E_{old})}$ nell'altro caso.

Nell'algoritmo di Gibbs, per ogni spin vengono calcolati $p_{up} = e^{-\beta E_{up}}$, $p_{down} = e^{-\beta E_{down}}$ e $p = \frac{p_{up}}{p_{up}+p_{down}}$. Indipendentemente dal valore originario dello spin, esso viene posto up con probabilità $p$. L'algoritmo di Gibbs è dunque un algoritmo di Metropolis con probabilità di transizione $T(up|down) = p$ non simmetrica e accettanza sempre uno.

L'esercizio richiede di inserire la possibilità di ripartire da una configurazione di spin precedente e di valutare l'energia interna per particella, la capacità termica per particella, la suscettività con campo esterno nullo e la magnetizzazione con $h = 0.02$, dopo aver lasciato equilibrare il sistema.

#### Risultati

Si sceglie di lasciar equilibrare il sistema per 500 step prima di operare la presa dati.

Le osservabili termodinamiche sono calcolate in questo modo:
$$U(N,T)/N = \langle H \rangle / N$$
$$C(N,T) k_{b}/N = \frac{1}{T^2}\left(\left\langle \frac{H^2}{N} \right\rangle - \left\langle \frac{H}{N} \right\rangle^2\right)$$
$$M(N,T) = \left\langle \sum_{i=1,N}s_i \right\rangle$$
$$\chi(N,T) k_b = \frac{1}{T} \left\langle \left(\sum_{i=1,N}s_i\right)^2 \right\rangle$$

Si riportano i grafici delle grandezze termodinamiche in funzione del numero di blocchi con campionamento di Gibbs e Metropolis per $T = 0.5$K. Si utilizzano $100$ blocchi con $1000$ step in ogni blocco.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

ax=ax1
heat = np.loadtxt("output.heat.g0.500000", usecols=2)
err_heat = np.loadtxt("output.heat.g0.500000", usecols=3)
ax.errorbar(N,heat,yerr=err_heat)
ax.set_xlabel('#blocchi')
ax.set_ylabel(r'$C/N$ $[k_{b}]$')
ax.set_title("Capacità termica per particella - Gibbs")
ax.grid(True)

ax=ax2
heat = np.loadtxt("output.heat.0.500000", usecols=2)
err_heat = np.loadtxt("output.heat.0.500000", usecols=3)
ax.errorbar(N,heat,yerr=err_heat, color='red')
ax.set_xlabel('#blocchi')
ax.set_ylabel(r'$C/N$ $[k_{b}]$')
ax.set_title("Capacità termica per particella - Metropolis")
ax.grid(True)
```

Vengono ora presentati i grafici delle quattro grandezze termodinamiche in funzione della temperatura in un range $[0.5; 2.0]$K per campionamento di Gibbs e di Metropolis, sovrapposti alle curve teoriche ottenute con uno studio analitico del modello di Ising unidimensionale. Gli andamenti teorici nei grafici sono di colore arancione. Le incertezze statistiche delle curve numeriche sono spesso così piccole da non essere visibili. I punti sulle curve numeriche si riferiscono alle temperature a cui si esegue la simulazione.

```python
x=[0.5, 0.65, 0.8, 0.95, 1.1, 1.25, 1.4, 1.55, 1.7, 1.85, 2]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

ax=ax1
ene = np.loadtxt("values.dat", usecols=2)
err_ene = np.loadtxt("values.dat", usecols=3)
ax.errorbar(x,ene,yerr=err_ene, marker='o')
ax.set_xlabel('$T$ [K]')
ax.set_ylabel(r'$U/N$ [J]')
ax.set_title("Energia interna in funzione della temperatura - Gibbs")
ax.grid(True)

points=100
T = np.linspace(0.5,2.,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
ax.plot(T, e, color='orange')

ax=ax2
ene = np.loadtxt("values_metro.dat", usecols=2)
err_ene = np.loadtxt("values_metro.dat", usecols=3)
ax.errorbar(x,ene,yerr=err_ene, color='red', marker='o')
ax.plot(T, e, color='orange')
ax.set_xlabel('$T$ [K]')
ax.set_ylabel(r'$U/N$ [J]')
ax.set_title("Energia interna in funzione della temperatura - Metropolis")
ax.grid(True)
```

Si nota una buona sovrapposizione tra modello teorico e risultato numerico. All'aumentare della temperatura è favorito il disallineamento degli spin perché il peso di Boltzmann tende ad uno e dunque la proposta di cambiamento dello spin viene accettata con alta probabilità. A bassa temperatura il sistema tende ad allinearsi, l'esponente $\Delta E$ nel peso di Boltzmann risulta prevalentemente positivo e dunque diminuisce l'accettanza. Ciò è evidente dall'*acceptance rate* stampato a schermo eseguendo il codice con Metropolis: l'accettanza assume valori compresi tra $0.02$ e $0.54$ circa passando da temperature basse ad alte.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

ax=ax1
heat = np.loadtxt("values.dat", usecols=5)
err_heat = np.loadtxt("values.dat", usecols=6)
ax.errorbar(x,heat,yerr=err_heat, marker='o')
ax.set_xlabel('$T$ [K]')
ax.set_ylabel(r'$C/N$ $[k_{b}]$')
ax.set_title("Capacità termica per particella in funzione della temperatura - Gibbs")
ax.grid(True)

h=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
ax.plot(T, h, color='orange')

ax=ax2
heat = np.loadtxt("values_metro.dat", usecols=5)
err_heat = np.loadtxt("values_metro.dat", usecols=6)
ax.errorbar(x,heat,yerr=err_heat, color='red', marker='o')
ax.plot(T, h, color='orange')
ax.set_xlabel('$T$ [K]')
ax.set_ylabel(r'$C/N$ $[k_{b}]$')
ax.set_title("Capacità termica per particella in funzione della temperatura - Metropolis")
ax.grid(True)
```

La capacità termica, derivata dell'energia interna rispetto alla temperatura, assume un valore massimo attorno a $0.8$K per poi diminuire sempre più verso lo zero per via del numero finito di gradi di libertà. La sovrapposizione tra modello teorico e risultati numerici non è perfetta per temperature $< 0.8$K.

```python
x=[0.5, 0.65, 0.8, 0.95, 1.1, 1.25, 1.4, 1.55, 1.7, 1.85, 2]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

ax=ax1
mag = np.loadtxt("valuesh.dat", usecols=8)
err_mag = np.loadtxt("valuesh.dat", usecols=9)
ax.errorbar(x,mag,yerr=err_mag, marker='o')
ax.set_xlabel('$T$ [K]')
ax.set_ylabel(r'$M$ [A/m]')
ax.set_title("Magnetizzazione in funzione della temperatura con $h = 0.02$ - Gibbs")
ax.grid(True)

h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
ax.plot(T, M, color='orange')

ax=ax2
mag = np.loadtxt("values_metroh.dat", usecols=8)
err_mag = np.loadtxt("values_metroh.dat", usecols=9)
ax.errorbar(x,mag,yerr=err_mag, color='red', marker='o')
plt.plot(T, M, color='orange')
ax.set_xlabel('$T$ [K]')
ax.set_ylabel(r'$M$ [A/m]')
ax.set_title("Magnetizzazione in funzione della temperatura con $h = 0.02$ - Metropolis")
ax.grid(True)
```

La magnetizzazione con campo esterno non nullo assume valori massimi per temperature basse per poi tendere a zero. Il sistema favorisce infatti una configurazione disordinata a temperature alte nonostante il campo esterno. Le curve ottenute sono ben confrontabili con quelle attese.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

ax=ax1
chi = np.loadtxt("values.dat", usecols=11)*50
err_chi = np.loadtxt("values.dat", usecols=12)*50
ax.errorbar(x,chi,yerr=err_chi, marker='o')
ax.set_xlabel('$T$ [K]')
ax.set_ylabel(r'$\chi$ [A^2/(K m^2) $k_{b}$]')
ax.set_title("Suscettività in funzione della temperatura - Gibbs")
ax.grid(True)

X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
ax.plot(T, X, color='orange')

ax=ax2
chi = np.loadtxt("values_metro.dat", usecols=11)*50
err_chi = np.loadtxt("values_metro.dat", usecols=12)*50
ax.errorbar(x,chi,yerr=err_chi, color='red', marker='o')
ax.plot(T, X, color='orange')
ax.set_xlabel('$T$ [K]')
ax.set_ylabel(r'$\chi$ [A^2/(K m^2) $k_{b}$]')
ax.set_title("Suscettività in funzione della temperatura - Metropolis")
ax.grid(True)
```

La suscettività diminuisce all'aumentare della temperatura perché inversamente proporzionale alla temperatura stessa. Le incertezze statistiche sui valori di suscettività sono nulli o a volte così piccoli da non essere rappresentabili dal calcolatore. I risultati numerici non si sovrappongono con quelli teorici.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

ax=ax1
chi = np.loadtxt("values_metro10.dat", usecols=5)
err_chi = np.loadtxt("values_metro10.dat", usecols=6)
ax.errorbar(x,chi,yerr=err_chi, color='red', marker='o')
ax.set_xlabel('$T$ [K]')
ax.set_ylabel(r'$C/N$ $[k_{b}]$')
ax.set_title("Capacità termica per particella in funzione della temperatura - Metropolis")
ax.grid(True)

h=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
ax.plot(T, h, color='orange')

ax=ax2
chi = np.loadtxt("values_metro10.dat", usecols=11)*50
err_chi = np.loadtxt("values_metro10.dat", usecols=12)*50
ax.errorbar(x,chi,yerr=err_chi, color='red', marker='o')
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
ax.plot(T, X, color='orange')
ax.set_xlabel('$T$ [K]')
ax.set_ylabel(r'$\chi$ [A^2/(K m^2) $k_{b}$]')
ax.set_title("Suscettività in funzione della temperatura - Metropolis")
ax.grid(True)
```
