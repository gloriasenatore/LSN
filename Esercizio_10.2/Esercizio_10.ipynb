### Esercizio 10.1

#### Svolgimento

Scopo dell'esercizio è scrivere un algoritmo di *simulated annealing* per risolvere il problema del commesso viaggiatore e confrontare i risultati ottenuti con quelli dell'esercitazione $9$.

Nella classe *Population* dell'esercitazione $9$ si inserisce il nuovo metodo *Move(vector G, double $\beta$)*, che sottopone l'individuo contenente le città a uno dei tre operatori di permutazioni descritti nell'esercitazione precedente, con probabilità del $33\%$ per ciascuno. Viene in seguito eseguita la mossa dell'algoritmo di Metropolis: si accetta il nuovo individuo con probabilità $e^{-\beta(L_{new} - L_{old})}$, con $L_{new}$ e $L_{old}$ lunghezze del percorso del nuovo individuo e dell'individuo originale e $\beta$ un numero che aumenta al progredire delle iterazioni. 

L'idea dell'algoritmo è infatti di operare l'*embedding* del problema con un sistema di meccanica statistica. Il sistema è inizialmente a temperatura elevata e le mosse dell'algoritmo di Metropolis vengono accettate con una certa probabilità. In questa situazione il sistema è ergodico e libero di esplorare diversi stati nello spazio dei parametri. Abbassando la temperatura, le permutazioni sull'individuo vengono accettate con minor probabilità, cioé viene esplorato uno spazio dei parametri più ristretto attorno all'individuo corrente, e l'algoritmo trova infine una configurazione che minimizza la funzione costo.

#### Risultati

### Esercizio 10.2

#### Svolgimento

Si modifica ora il codice dell'esercizio $9$ inserendo la possibilità di utilizzare più *core* di calcolo tramite programmazione parallela. Dopo aver incluso la libreria MPI, e aver definito il significato di *size*, numero di processi utilizzati in tutto, e *rank*, numero del singolo processo, ogni continente performa le iterazioni dell'algoritmo genetico in maniera indipendente. Si noti che nella classe *Population* ogni processo ha un seme del generatore di numeri casuali differente dagli altri. 

A partire dall'iterazione $0$ e in seguito ogni $200$ iterazioni, i quattro continenti si scambiano l'individuo migliore, cioè l'individuo con il miglior *fitness* fino a quel momento, casualmente. Il *rank* $0$ decide in maniera casuale a quale continente ciascun *rank* deve inviare il proprio individuo migliore. Sempre il processo $0$ invia l'informazione a tutti gli altri tramite un messaggio di tipo *broadcast*. Ciascun processo procede con il trasferimento del proprio individuo migliore con una chiamata *MPI_Isend*, ossia non bloccante. Il processo ricevente acquisisce con la subroutine bloccante *MPI_Recv* l'individuo e lo sostituisce al proprio miglior individuo. In questo modo, gli individui migliori hanno tutti cambiato continente. Senza riordinare la popolazione, in modo che ci sia probabilità alta di scegliere il migliore appena acquisito per un eventuale *crossover*, i *rank* procedono con una iterazione dell'algoritmo genetico, e poi con altre $199$ iterazioni normalmente. 

#### Risultati

Si riportano i risultati ottenuti per città distribuite sulla superficie del quadrato utilizzando una popolazione di $200$ individui in ogni continente e iterando per $20000$ volte l'algoritmo genetico, con trasferimento dell'individuo migliore ogni $200$ iterazioni.

```python
plt.figure(figsize=(7,7))
x = np.loadtxt("bestpath_rank0.dat", usecols=0)
y = np.loadtxt("bestpath_rank0.dat", usecols=1)
c = np.loadtxt("bestpath_rank0.dat", usecols=2)
L = np.loadtxt("bestpath_rank0.dat", usecols=3, max_rows=1)
t = np.linspace(-1,1,100)
plt.plot(x,y, color='red')
for i, txt in enumerate(c):
    plt.annotate(txt, (x[i], y[i]))
plt.hlines(-1, -1, 1, color='blue')
plt.hlines(1, -1, 1, color='blue')
plt.vlines(-1, -1, 1, color='blue')
plt.vlines(1, -1, 1, color='blue')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title("Percorso migliore - città sulla superficie del quadrato - Algoritmo genetico in programmazione parallela")
plt.grid(False)
plt.show()
print("Lunghezza del percorso migliore:")
print(L)
```

La lunghezza del percorso migliore alla fine di tutte le iterazioni è uguale a quella ottenuta nell'esercitazione $9$. La figura in alto è relativa al processo numero $0$, ma si riportano tra i file di output anche i *bestpath* degli altri *rank*, da cui si nota che tutti i processi giungono allo stesso risultato. Il percorso trovato è tuttavia di sicuro un minimo locale, infatti l'algoritmo che performa il *simulated annealing* trova un percorso migliore.

```python
plt.figure(figsize=(8,6))
L0 = np.loadtxt("bestL_rank0.dat")
L1 = np.loadtxt("bestL_rank1.dat")
L2 = np.loadtxt("bestL_rank2.dat")
L3 = np.loadtxt("bestL_rank3.dat")
x = np.arange(0,20001)

plt.plot(x, L0, label='rank 0')
plt.plot(x, L1, color='red', label = 'rank 1')
plt.plot(x, L2, color='green', label = 'rank 2')
plt.plot(x, L3, color='black', label = 'rank 3')
plt.xlabel('iteration')
plt.ylabel('$L$')
plt.title("Lunghezza del percorso in funzione dell'iterazione per i quattro rank - Algoritmo genetico")
plt.legend(loc=(1.1, 0.8), frameon=True)
plt.grid(True)
plt.show()
```

Si grafica la lunghezza del percorso migliore in funzione dell'iterazione per i quattro continenti. Confrontando con lo stesso grafico dell'esercitazione $9$, si nota che le performance dell'algoritmo sono molto migliorate, infatti tutti i *rank* trovano l'individuo migliore dopo già $6000$ iterazioni circa, mentre nell'altra esercitazione un solo *rank* impiega più di $16000$ iterazioni.

```python
plt.figure(figsize=(8,6))
L0 = np.loadtxt("bestL_rank0.dat", max_rows=1000)
L1 = np.loadtxt("bestL_rank1.dat", max_rows=1000)
L2 = np.loadtxt("bestL_rank2.dat", max_rows=1000)
L3 = np.loadtxt("bestL_rank3.dat", max_rows=1000)
x = np.arange(0,1000)

plt.plot(x, L0, label='rank 0')
plt.plot(x, L1, color='red', label = 'rank 1')
plt.plot(x, L2, color='green', label = 'rank 2')
plt.plot(x, L3, color='black', label = 'rank 3')
plt.xlabel('iteration')
plt.ylabel('$L$')
plt.title("Lunghezza del percorso in funzione dell'iterazione per i quattro rank - Algoritmo genetico")
plt.legend(loc=(1.1, 0.8), frameon=True)
plt.grid(True)
plt.show()
```

Con un zoom sulle prime $1000$ iterazioni si osserva che avvengono vistose diminuzioni della lunghezza del percorso migliore in prossimità di multipli di $200$ iterazioni. Probabilmente esse sono connesse allo scambio degli individui migliori. Alla fine delle $1000$ iterazioni tutti i continenti concordano sul fatto che l'individuo migliore abbia lunghezza minore di $10$ in norma quadra, mentre nell'esercitazione $9$ dopo lo stesso numero di iterazioni l'individuo migliore ha ancora lunghezza maggiore di $10$.
