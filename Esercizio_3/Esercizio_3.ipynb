### Esercizio 3.1

#### Svolgimento

L'esercizio richiede di stimare con metodi Monte Carlo il prezzo al tempo iniziale di opzioni europee di tipo *call* e *put*. 

Sia $r$ il tasso d'interesse a rischio zero, ossia il tasso d'interesse garantito da una banca per il deposito, e $\sigma$ la volatilità, ovvero la misura delle fluttuazioni della parte statistica del prezzo dell'opzione. Si procede in due modi. Dato il prezzo iniziale del sottostante $S(0)$, si campiona direttamente il prezzo finale, avendo quest'ultimo la forma di un moto browniano geometrico di drift $r$ e coefficiente di diffusione $\sigma^2$, sfruttando la legge: 

$$S(T) = S(0) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)T+\sigma Z \sqrt{T}\right]$$
con $Z$ tratto da una distribuzione gaussiana di media $0$ e varianza $1$ e con $T$ l'*expiry date*.

Il secondo modo prevede di campionare il prezzo del sottostante discretizzato dividendo l'intervallo $[0,T]$ in cento sottointervalli con la relazione:

$$S(t_{i+1}) = S(t_{i}) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)(t_{i+1} - t_{i})+\sigma Z_{i+1} \sqrt{t_{i+1} - t_{i}}\right]$$
per $t_{i}$ che va da $t_{0} = 0$ a $t_{100} = T$.

In entrambi i casi si prosegue valutando il prezzo al tempo iniziale dell'opzione *call* e *put* rispettivamente come:
$$C = exp(-rT) \max({0, S(T)-K)}$$
$$P = exp(-rT) \max({0, K-S(T))}$$
dove il termine esponenziale è il fattore da scontare per via degli interessi garantiti da una banca per il deposito al tempo iniziale, in modo che al tempo finale il profitto dell'opzione *call* sia proprio $\max({0, S(T)-K)}$ e *put* $\max({0, K-S(T))}$.

Si ripete il calcolo per $L = 1000$ volte in ognuno degli $N = 100$ blocchi e si valutano infine medie progressive ed incertezze statistiche.

#### Risultati

Si riportano i grafici delle medie progressive dei prezzi di opzioni *call* e *put* con le relative incertezze statistiche. Di fianco ad ogni grafico, inoltre, si plottano le differenze tra risultati trovati e risultati analitici previsti da Black e Scholes.

Risultati per il campionamento diretto:

```python
M=100000              
N=100                 
L=int(M/N)
x = np.arange(N)

C = np.loadtxt("C_diretto")
err_C = np.loadtxt("C_diretto_err")
Black_Scholes = np.loadtxt("Black-Scholes.txt")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

ax=ax1
ax.errorbar(x,C,yerr=err_C)
ax.set_xlabel('#blocchi')
ax.set_ylabel(r'$<C>$')
ax.set_title("Stima del prezzo $C$ della $\it{call}$ con campionamento diretto")
ax.grid(True)

ax=ax2
ax.errorbar(x,C-Black_Scholes[0],yerr=err_C)
ax.set_xlabel('#blocchi')
ax.set_ylabel(r'$<C> - C_{BS}$')
ax.set_title("Confronto con il risultato analitico di Black e Scholes")
ax.grid(True)
```

```python
P = np.loadtxt("P_diretto")
err_P = np.loadtxt("P_diretto_err")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

ax=ax1
ax.errorbar(x,P,yerr=err_P, color='red')
ax.set_xlabel('#blocchi')
ax.set_ylabel(r'$<P>$')
ax.set_title("Stima del prezzo $P$ della $\it{put}$ con campionamento diretto")
ax.grid(True)

ax=ax2
ax.errorbar(x,P-Black_Scholes[1],yerr=err_P, color='red')
ax.set_xlabel('#blocchi')
ax.set_ylabel(r'$<P> - P_{BS}$')
ax.set_title("Confronto con il risultato analitico di Black e Scholes")
ax.grid(True)
```

Risultati per il campionamento discreto:

```python
C = np.loadtxt("C_discreto")
err_C = np.loadtxt("C_discreto_err")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

ax=ax1
ax.errorbar(x,C,yerr=err_C)
ax.set_xlabel('#blocchi')
ax.set_ylabel(r'$<C>$')
ax.set_title("Stima del prezzo $C$ della $\it{call}$ con campionamento discreto")
ax.grid(True)

ax=ax2
ax.errorbar(x,C-Black_Scholes[0],yerr=err_C)
ax.set_xlabel('#blocchi')
ax.set_ylabel(r'$<C> - C_{BS}$')
ax.set_title("Confronto con il risultato analitico di Black e Scholes")
ax.grid(True)
```

```python
P = np.loadtxt("P_discreto")
err_P = np.loadtxt("P_discreto_err")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

ax=ax1
ax.errorbar(x,P,yerr=err_P, color='red')
ax.set_xlabel('#blocchi')
ax.set_ylabel(r'$<P>$')
ax.set_title("Stima del prezzo $P$ della $\it{put}$ con campionamento discreto")
ax.grid(True)

ax=ax2
ax.errorbar(x,P-Black_Scholes[1],yerr=err_P, color='red')
ax.set_xlabel('#blocchi')
ax.set_ylabel(r'$<P> - P_{BS}$')
ax.set_title("Confronto con il risultato analitico di Black e Scholes")
ax.grid(True)
```

Sia con il metodo del campionamento diretto che con quello del campionamento discreto si ottengono prezzi delle opzioni compatibili con i risultati analitici di Black e Scholes, con le incertezze statistiche che diminuiscono all'aumentare del numero di tiri. Si osserva che con la discretizzazione del prezzo finale ognuna delle $M = 10^5$ simulazioni produce un prezzo statisticamente compatibile con quello trovato campionando direttamente, anche se non esattamente uguale.
