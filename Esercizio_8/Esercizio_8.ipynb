### Esercizio 8.1

#### Svolgimento

L'esercizio richiede di implementare un algoritmo Monte Carlo per lo studio dello stato fondamentale di una particella quantistica unidimensionale immersa nel potenziale $V(x) = x^4 - \frac{5}{2}x^2$ con metodo variazionale.
Lo stato fondamentale tentativo è $$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}} + e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$ Esso può essere considerato un buon *ansatz* del vero stato fondamentale sconosciuto, essendo simmetrico rispetto all'origine, come il potenziale, e costituito da due gaussiane centrate in $\mu$ e $-\mu$, allo stesso modo in cui il potenziale ha due minimi simmetrici rispetto all'origine.

Per ottenere il valore d'aspettazione dell'hamiltoniana $$
\langle {\hat H} \rangle_T = \int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$ su questo stato tentativo, si implementa un algoritmo di Metropolis. Partendo da $x$ posto nell'origine, si genera un passo con probabilità di transizione $T(x_{new}|x_{old})$ uniforme e si valuta l'accettanza come $min(1, \frac{|\Psi_{T}(x_{new})|^2}{|\Psi_{T}(x_{old})|^2})$, essendo $T$ simmetrica ed essendo la normalizzazione della densità di probabilità uguale per numeratore e denominatore. Con i parametri $\sigma = 0.6$ e $\mu = 0.8$, si osserva che il passo deve essere ampio circa $5$ per avere un'accettanza del $50\%$. 

L'energia locale $\frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}$ si valuta come $$
\frac{{(-\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2} + V(x))}\Psi_T^{\sigma,\mu}(x)}{\Psi_T^{\sigma,\mu}(x)} =
\dfrac{\frac{1}{2\sigma^4} e^{-\frac{(x+\mu)^2}{2\sigma^2}}(e^{\frac{2x\mu}{\sigma^2}}(\mu-\sigma-x)(\mu+\sigma-x) + (\mu+x)^2-\sigma^2)}{\Psi_T^{\sigma,\mu}(x)} + V(x)
$$
per $\hbar = 1$ e $m = 1$.

#### Risultati

Eseguendo il *data blocking* con $N=100$ blocchi e $L=10^4$ simulazioni in ogni blocco, con $\sigma = 0.6$ e $\mu = 0.8$, si ottiene il seguente grafico:

```python
x = np.arange(100)
H = np.loadtxt("8.1.H_medio")
H_err = np.loadtxt("8.1.H_err")

plt.figure(figsize=(8,5))
plt.errorbar(x,H,yerr=H_err)
plt.xlabel('#blocchi')
plt.ylabel(r'$\langle {\hat H} \rangle_T$')
plt.title("Valor medio dell'hamiltoniana sullo stato tentativo, $\sigma = 0.6$, $\mu = 0.8$")
plt.grid(True)
plt.show()
```

Le incertezze statistiche diminuiscono all'aumentare del numero di blocchi e l'energia è abbastanza vicina a quella ottenuta risolvendo l'equazione di Schrodinger con metodo matriciale, pari a circa $-0.46$. I valori scelti per $\sigma$ e $\mu$ sono dunque già molto vicini a quelli che minimizzano il valore dell'energia sullo stato fondamentale tentativo.

Il valore energetico dello stato fondamentale trovato è dunque $-0.4440 \pm 0.0008$. Poiché si è utilizzato un approccio variazionale, si sa che tale valore è leggermente superiore al vero *ground state*.

Si riporta ora il grafico del potenziale e della densità di probabilità $|\Psi_{T}(x)|^2$ campionata per i valori di $\sigma$ e $\mu$ del grafico precedente, da confrontare con la curva analitica.

```python
x1 = np.loadtxt("8.2.histo.dat", usecols=0)
psi2 = np.loadtxt("8.2.histo.dat", usecols=1)

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

def Vpot(x):
    return (x**2 - 2.5)*x**2

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

def Vpot(x):
    return (x**2 - 2.5)*x**2
scale = 0.3

plt.figure(figsize=(8,5))
plt.plot(x,(psi[0])**2, label='${|\psi|}^2$ matriciale')
plt.plot(x1,scale*psi2, label='${|\psi_{T}|}^2$ variazionale')
plt.plot(x1,scale*Vpot(x1), label='$V(x)$', color='Black')
plt.ylim(-0.6, 0.6)
plt.xlim(-3,3)
plt.xlabel('$x$')
plt.title("Potential & Probabilities")
plt.grid(True)
plt.legend(loc=(1.05, 0.75), frameon=True)
plt.show()
```

La curva ottenuta con metodo numerico è simile a quella ottenuta con approccio matriciale. In corrispondenza dei due minimi del potenziale si trovano i massimi di $|\psi_{T}|^2$ e, avvicinandosi a $x=0$, dove il potenziale ha un massimo locale, la densità di probabilità della funzione d'onda ha un minimo locale.
