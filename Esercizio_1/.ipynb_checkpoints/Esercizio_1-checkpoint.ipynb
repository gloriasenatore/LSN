### Esercizio 1.1

#### Svolgimento

Per stimare il valor medio di $r$, si generano $L$ numeri uniformemente distribuiti tra 0 e 1 in ogni blocco. Si procede calcolando la media e il valor medio quadro di tali numeri in ogni blocco. Valutare la media cumulativa e la sua incertezza permette di controllare l'andamento di $<r>$ e valutarne la convergenza al valore atteso.

L'ultimo punto dell'esercizio 1.1 si riferisce al test del $\chi^2$ con cui si intende verificare che la probabilità che i numeri ottenuti dal generatore di numeri pseudocasuali non siano uniformemente distribuiti sia bassa. Tale metodo può essere dunque un mezzo per accertarsi della bontà del proprio generatore. Per cento volte vengono generati $10^4$ numeri interi uniformemente distribuiti tra 0 e 100 e si conta quanti numeri cadono in ognuno dei cento sottointervalli di $[0, 100]$. Se $\chi^2$ vale circa 100, l'ipotesi che i numeri siano tratti da una distribuzione uniforme è accettabile.

#### Risultati

Si plotta $<r> - 1/2$ al crescere del numero di blocchi.

```python
M=100000       #Numero totale throws       
N=100          #Numero blocchi       
L=int(M/N)     #throws in ogni blocco
x = np.arange(N)
```

```python
plt.figure(figsize=(8,5))
sum_prog = np.loadtxt("Valor_medio_1.1")
err_r = np.loadtxt("Valor_medio_err_1.1")
plt.errorbar(x,sum_prog-0.5,yerr=err_r)
plt.xlabel('#blocchi')
plt.ylabel(r'$<r>-1/2$')
plt.title("Stima di <r> in funzione del numero di blocchi")
plt.grid(True)
plt.show()
```

Il grafico mostra come il valor medio di $<r>$ tenda al valore atteso entro errori statistici. Per il calcolo di tali errori vengono sfruttate le conseguenze del teorema del limite centrale, ossia il risultato secondo cui la distribuzione di probabilità della media di variabili statisticamente indipendenti ha una deviazione standard pari alla deviazione standard delle variabili diviso il numero di variabili. 

Le variabili pseudocasuali generate hanno tuttavia un certo grado di correlazione. Ciò che giustifica quindi l'utilizzo del teorema del limite centrale è il raggruppamento a blocchi. Se il numero di blocchi $N$ è molto maggiore della distanza di correlazione tra le variabili, infatti, calcolare il valor medio di ogni blocco abbatte la correlazione tra valori medi.

Per valutare la deviazione standard di variabili uniformemente distribuiti da $0$ e $1$ viene ora graficato $<r - 1/2>^2 - 1/12$.

```python
plt.figure(figsize=(8,5))
sum_prog_var = np.loadtxt("Valor_medio_var_1.1")
err_var = np.loadtxt("Valor_medio_var_err_1.1")
plt.errorbar(x,sum_prog_var-(1/12),yerr=err_var)
plt.xlabel('#blocchi')
plt.ylabel(r'$<r-1/2>^2-1/12$')
plt.title("Stima di $\sigma^2$ in funzione del numero di blocchi")
plt.grid(True)
plt.show()
```

Si osserva come la varianza tenda al valore atteso di $1/12$.

Vengono plottati i valori del ${\chi_{j}}^2$ al variare della j-esima simulazione.

```python
plt.figure(figsize=(8,5))
j = np.arange(100)
chi2 = np.loadtxt("Chi2_1.1")
plt.plot(j, chi2)
plt.xlabel('j')
plt.ylabel(r'$\chi_{j}^2$')
plt.title("Valori di $\chi^2$ ottenuti nelle 100 simulazioni")
plt.grid(True)
plt.show()
```

I valori ottenuti variano attorno a 100 come atteso. Le fluttuazioni sono tuttavia abbastanza ampie. Ciò è conseguenza del fatto che i generatori pseudocasuali, poiché basati su algoritmi deterministici, mantengono una certa correlazione tra le variabili. Il generatore utilizzato è inoltre lineare congruenziale e nasconde dunque correlazioni spaziali tra le variabili. Utilizzando algoritmi di *reshuffling* si evita di generare numeri distribuiti su iperpiani, superando il problema delle correlazioni spaziali.

### Esercizio 1.2

#### Svolgimento

In questo esercizio vengono studiate tre distribuzioni di probabilità, uniforme, esponenziale e lorentziana e viene verificato il teorema del limite centrale. Per $N = 1, 2, 10, 100$ si calcola $10^4$ volte il valor medio di numeri generati con le tre distribuzioni. Per generare numeri dati con una certa distribuzione di probabilità si sfrutta il metodo dell'inversione della cumulativa, dato che questa è conosciuta e invertibile per distribuzione esponenziale e lorentziana. Data $y$ uniformemente distribuita in $[0, 1)$, queste sono le relazioni per ottenere $x$ rispettivamente distribuita in modo esponenziale e lorentziano:

$$x = -\frac{1}{\lambda}\ln{(1-y)}$$

$$x = \Gamma \tan(\pi(y-\frac{1}{2})) + \mu$$


Per $N = 100$ si fittano con una gaussiana gli istogrammi della distribuzione uniforme ed esponenziale e con una lorentziana l'istogramma della distribuzione lorentziana. Alla varianza delle gaussiane si assegna il valore di $n - \mu$, con $n$ il valore del numero entro cui si trovano il 68.3% dei numeri maggiori di $\mu$, valor medio delle gaussiane.

#### Risultati

Di seguito si riportano gli istogrammi delle tre distribuzioni per i quattro valori di $N$.

```python
std_1 = np.loadtxt("Risultati_std_1.2", max_rows = 9999)
std_2 = np.loadtxt("Risultati_std_1.2", skiprows = 9999, max_rows = 19999)
std_10 = np.loadtxt("Risultati_std_1.2", skiprows = 19999, max_rows = 29999)
std_100 = np.loadtxt("Risultati_std_1.2", skiprows = 29999, max_rows = 39999)

fig, axx = plt.subplots(2, 2, figsize=(15,11))
axx = axx.reshape(-1)
arguments = [std_1, std_2, std_10, std_100]
Ns = [1, 2, 10, 100]
i = 0
for argument in arguments:
        N = Ns[i]
        ax = axx[i]
        M = max(argument)
        m = min(argument)
        ax.hist(argument, 100, range=(m, M), density=True)
        ax.set_title(f"Dado standard N = {N}")
        ax.set_xlabel('bins')
        ax.set_ylabel('dati')
        ax.grid(True)
        i+=1

mu = 0.5
##Stima del valore di sigma
number = []
std = []
n = 0
for std in std_100:
    if(std > mu):
        number.append(std)
        n = n+1
number = sorted(number)
        
sigma = number[int(0.683*n)] - 0.5
bins = np.linspace(m,M,100)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(- (bins - mu)**2 / (2 * sigma**2)), 
         linewidth=2, color='r', label="Fit gaussiano")
plt.legend(loc=(0.7, 0.8), frameon=True)
plt.show()
```

Si osserva come, per la distribuzione uniforme, i numeri uniformemente distribuiti tra $0$ e $1$ si distribuiscano in modo sempre più denso attorno a $0.5$ al crescere di N. Con $N = 100$ la distribuzione dei numeri può essere fittata con una gaussiana con media $\mu$. Ciò costituisce una verifica del teorema del limite centrale, infatti le variabili utilizzate sono identicamente distribuite (in realtà non è una condizione necessaria), in buona approssimazione statisticamente indipendenti e a varianza finita. 

```python
exp_1 = np.loadtxt("Risultati_exp_1.2", max_rows = 9999)
exp_2 = np.loadtxt("Risultati_std_1.2", skiprows = 9999, max_rows = 19999)
exp_10 = np.loadtxt("Risultati_std_1.2", skiprows = 19999, max_rows = 29999)
exp_100 = np.loadtxt("Risultati_std_1.2", skiprows = 29999, max_rows = 39999)

fig, axx = plt.subplots(2, 2, figsize=(15,11))
axx = axx.reshape(-1)
arguments = [exp_1, exp_2, exp_10, exp_100]
i = 0
for argument in arguments:
        N = Ns[i]
        ax = axx[i]
        M = max(argument)
        m = min(argument)
        ax.hist(argument, 100, range=(m, M), density=True)
        ax.set_title(f"Dado esponenziale N = {N}")
        ax.set_xlabel('bins')
        ax.set_ylabel('dati')
        ax.grid(True)
        i+=1
        
mu = 0.5
number = []
std = []
n = 0
for exp in exp_100:
    if(exp > mu):
        number.append(exp)
        n = n+1
number = sorted(number)
        
sigma = number[int(0.683*n)] - 0.5
bins = np.linspace(m,M,100)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp(- (bins - mu)**2 / (2 * sigma**2)), 
         linewidth=2, color='r', label="Fit gaussiano")
plt.legend(loc=(0.7, 0.8), frameon=True)
plt.show()
```

Anche nel caso di numeri tratti da distribuzione esponenziale con $\lambda = 1$ si osserva al crescere di $N$ convergenza verso una distribuzione gaussiana con media $0.5$. La distribuzione gaussiana è infatti un attrattore nello spazio delle distribuzioni di probabilità a patto che siano rispettate le condizioni per cui è verificato il teorema del limite centrale.

```python
lor_1 = np.loadtxt("Risultati_lor_1.2", max_rows = 9999)
lor_2 = np.loadtxt("Risultati_lor_1.2", skiprows = 9999, max_rows = 19999)
lor_10 = np.loadtxt("Risultati_lor_1.2", skiprows = 19999, max_rows = 29999)
lor_100 = np.loadtxt("Risultati_lor_1.2", skiprows = 29999, max_rows = 39999)

fig, axx = plt.subplots(2, 2, figsize=(15,11))
axx = axx.reshape(-1)
arguments = [lor_1, lor_2, lor_10, lor_100]
i = 0
for argument in arguments:
        N = Ns[i]
        ax = axx[i]
        ax.hist(argument, 100, range=(-20, 20), density=True)
        ax.set_title(f"Dado lorentziano N = {N}")
        ax.grid(True)
        i+=1
        
mu = 0
Gamma = 1
bins = np.linspace(-20,20,100)
plt.plot(bins, 1/np.pi * Gamma/((bins-mu)**2 + Gamma**2), 
         linewidth=2, color='r', label="Fit lorentziano")
plt.legend(loc=(0.7, 0.8), frameon=True)
plt.show()
```

Con numeri ricavati da distribuzione lorentziana con $\Gamma = 1$ e $\mu = 0$ si osserva una rapida convergenza dei valori medi alla distribuzione lorentziana, che costituisce un attrattore alternativo alla gaussiana, valido ad esempio per distribuzione il cui secondo momento diverge.

### Esercizio 1.3

#### Svolgimento

Per simulare l'esperimento di Buffon, data la lunghezza dell'ago $L$ e la spaziatura tra le righe $d$, si genera $x$ uniformemente distribuito tra 0 e $d/2$ e $\theta$ tra 0 e $\pi/2$ e si incrementa il numero di volte in cui l'ago interseca la riga se $x \leq L\sin{\theta}$. Con $M = 10^7$ e $N = 100$ si ripete l'esperimento $F = M/N$ volte e si valutano i risultati medi e l'incertezza con il metodo di raggruppamento a blocchi. Il valore di $\pi$ in ogni blocco viene stimato con:

$$\begin{equation}
    \pi = \frac{2LF}{N_{hit}d}
\end{equation}$$ 

Per evitare di usare $\pi$ nell'algoritmo si utilizzano due metodi. Il primo è ricorsivo: partendo da, ad esempio, $p = 2$, $\theta$ viene generato tra 0 e $p/2$. Ogni volta il valore di $p$ viene aggiornato con la formula scritta sopra sostituendo al posto di $F$ il numero dell'esperimento che si sta eseguendo.

Il secondo metodo è di *reject*. Sono generati $x$ e $y$ tra $0$ e $1$. Se $x^2+y^2 < 1$, viene posto $\theta = \arccos{\frac{x}{\sqrt{x^2+y^2}}}$, altrimenti vengono generati altri numeri. 

#### Risultati

Vengono graficati le somme progressive del valore medio stimato meno il valore esatto di $\pi$ con le relative incertezze statistiche per i due metodi.

```python
M=10000000              
N=100                 
F=int(M/N)
x = np.arange(N)

plt.figure(figsize=(8,5))
sum_prog_p = np.loadtxt("Somme_progressive_p")
err_p = np.loadtxt("Err_p")
plt.errorbar(x,sum_prog_p-np.pi,yerr=err_p)
plt.xlabel('#blocchi')
plt.ylabel(r'$<p>-\pi$')
plt.title("Stima di $\pi$ metodo ricorsivo")
plt.grid(True)
plt.show()
```

Con il metodo ricorsivo si osserva un buon andamento dei dati al valore atteso entro le incertezze statistiche. Il metodo ricorsivo presenta tuttavia alcuni problemi. Per i primi esperimenti all'interno di ogni blocco la relazione scritta sopra non è una buona stima di $\pi$ perché è valida nel limite di numero di tiri alto, mentre al contrario sono stati eseguiti ancora pochi esperimenti. Un altro problema è legato al fatto che l'algoritmo dipende proprio dalla relazione di cui si vuole valutare la correttezza. Non conoscendo per nulla il valore di $\pi$, infine, ci si potrebbe chiedere se qualsiasi $p$ iniziale dia un risultato convergente a quello atteso. Provando con $p = 0.1$ e $p = 3$ si ottengono i seguenti risultati:

```python
sum_prog_p1 = np.loadtxt("Somme_progressive_p_0.1")
err_p1 = np.loadtxt("Err_p_0.1")
sum_prog_p2 = np.loadtxt("Somme_progressive_p_3")
err_p2 = np.loadtxt("Err_p_3")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

ax=ax1
ax.errorbar(x,sum_prog_p1-np.pi,yerr=err_p1)
ax.set_xlabel('#blocchi')
ax.set_ylabel(r'$<p>-\pi$')
ax.set_title("Stima di $\pi$ metodo ricorsivo $p = 0.1$")
ax.grid(True)

ax=ax2
ax.errorbar(x,sum_prog_p2-np.pi,yerr=err_p2)
ax.set_xlabel('#blocchi')
ax.set_ylabel(r'$<p>-\pi$')
ax.set_title("Stima di $\pi$ metodo ricorsivo $p = 3$")
ax.grid(True)
```

Si osserva che anche partendo da valori diversi di $p$ si ottiene una buona stima di $\pi$.

```python
plt.figure(figsize=(8,5))
sum_prog_p = np.loadtxt("Somme_progressive_p_rej")
err_p = np.loadtxt("Err_p_rej")
plt.errorbar(x,sum_prog_p-np.pi,yerr=err_p)
plt.xlabel('#blocchi')
plt.ylabel(r'$<p>-\pi$')
plt.title("Stima di $\pi$ metodo reject")
plt.grid(True)
plt.show()
```

Con il metodo *reject* si ottiene una abbastanza buona convergenza al valore atteso. In questo modo si evitano i problemi del primo metodo, anche se tuttavia si aggiunge il difetto che alcuni dei punti generati vengono scartati, essendo un metodo di *reject*.
